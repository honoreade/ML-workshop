```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
titles = pd.read_csv("google_books_dataset.csv")
titles.head(20)
```

Drop rows missing authors values:

#DataFrame.dropna(axis=0, how=’any’, thresh=None, subset=None, inplace=False)

```python
# title_clean = titles.dropna(axis= 0,subset = ["authors"])
# title_clean = title_clean.fillna('')
title_clean = titles.fillna('')
title_clean.shape
```

```python
book_content = title_clean['title'] +' '+ title_clean["authors"] +' '+ title_clean["categories"] +' '+ title_clean["publisher"]
book_content.head()
```

```python
# converting the text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
book_vectors = vectorizer.fit_transform(book_content)
# book_vectors = pd.DataFrame(book_vectors)
print(book_vectors)
```

```python
type(book_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(book_vectors)
similarity
```

```python
# getting the book name from the user

book_name = input('Enter your favourite book name : ')
```

```python
# creating a list with all the movie names given in the dataset

list_of_all_titles = title_clean['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(book_name, list_of_all_titles)
print(find_close_match)
```
